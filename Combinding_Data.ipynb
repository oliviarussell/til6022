```python
# create different lists per year
list2019 = [df2019Jan, df2019Feb, df2019Mar, df2019Apr, df2019May, df2019Jun, df2019Jul, df2019Aug, df2019Sep, df2019Oct, df2019Nov, df2019Dec]
list2020 = [df2020Jan, df2020Feb, df2020Mar, df2020Apr, df2020May, df2020Jun, df2020Jul, df2020Aug, df2020Sep, df2020Oct, df2020Nov, df2020Dec]
list2021 = [df2021Jan, df2021Feb, df2021Mar, df2021Apr, df2021May, df2021Jun, df2021Jul, df2021Aug, df2021Sep, df2021Oct, df2021Nov, df2021Dec]
years_list = [list2019, list2020, list2021] # sum up the lists so they can be found easily
```

```python
# create empty df's to store the data from before and post COVID
preCOVID = pd.DataFrame()
postCOVID = pd.DataFrame()

# loop through both the years list and the months to determine in which df each value should be.
for i in range(3):
    year = years_list[i]
    for k in range(12): #in here year[k] is the dataframe for a single month
        year[k]["Date"] = pd.to_datetime(year[k]["Date"])
        preCOVID_new = year[k][year[k]["Date"] < '2020-03-01'] #check for dates which are before March of 2020 and add them to a new dataframe
        preCOVID = pd.concat([preCOVID, preCOVID_new]) #add new dates from this iteration to the total df
        postCOVID_new = year[k][year[k]["Date"] > '2020-03-31'] #repeat two steps above for post-COVID
        postCOVID = pd.concat([postCOVID, postCOVID_new])

print(preCOVID)
print(postCOVID) #check files
```
