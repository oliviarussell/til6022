```python
# create different lists per year
list2019 = [df2019Jan, df2019Feb, df2019Mar, df2019Apr, df2019May, df2019Jun, df2019Jul, df2019Aug, df2019Sep, df2019Oct, df2019Nov, df2019Dec]
list2020 = [df2020Jan, df2020Feb, df2020Mar, df2020Apr, df2020May, df2020Jun, df2020Jul, df2020Aug, df2020Sep, df2020Oct, df2020Nov, df2020Dec]
list2021 = [df2021Jan, df2021Feb, df2021Mar, df2021Apr, df2021May, df2021Jun, df2021Jul, df2021Aug, df2021Sep, df2021Oct, df2021Nov, df2021Dec]
years_list = [list2019, list2020, list2021] # sum up the lists so they can be found easily
```

```python
from collections import Counter
#check for random years
# Initialize a Counter to hold year counts
year_counter = Counter()

# Loop through each year's list of DataFrames
for year_data in years_list:
    for df in year_data:
        # Convert the 'Date' column to datetime format
        df['Date'] = pd.to_datetime(df['Date'])
        
        # Extract the year from the Date column and update the counter
        years = df['Date'].dt.year
        year_counter.update(years)

# Print the counts of each year
print(year_counter)
```

```python
# create empty df's to store the data from before and post COVID
preCOVID = pd.DataFrame()
postCOVID = pd.DataFrame()

# loop through both the years list and the months to determine in which df each value should be.
for i in range(3):
    year = years_list[i]
    for k in range(12): #in here year[k] is the dataframe for a single month
        year[k]["Date"] = pd.to_datetime(year[k]["Date"])
        preCOVID_new = year[k][year[k]["Date"] < '2020-03-01'] #check for dates which are before March of 2020 and add them to a new dataframe
        preCOVID = pd.concat([preCOVID, preCOVID_new]) #add new dates from this iteration to the total df
        postCOVID_new = year[k][year[k]["Date"] > '2020-03-31'] #repeat two steps above for post-COVID
        postCOVID = pd.concat([postCOVID, postCOVID_new])

print(preCOVID)
print(postCOVID) #check files
```

```python
# Filter the data to remove all values which are outside of the range as they are apperentaly in the datasets
preCOVID = preCOVID[(preCOVID['Date'] > '2018-12-31') & (preCOVID['Date'] < '2022-01-01')]
postCOVID = postCOVID[(postCOVID['Date'] > '2018-12-31') & (postCOVID['Date'] < '2022-01-01')]
```

```python
# Calculate average number of trips per month in a function

def trips_per_month(df):
    df["Year_Month"] = df["Date"].dt.to_period("M")
    monthlyTrips = df.groupby("Year_Month").size()
    monthlyMean = monthlyTrips.mean()
    return monthlyMean, monthlyTrips

preCOVID_avg, preCOVID_monthly = trips_per_month(preCOVID)
postCOVID_avg, postCOVID_monthly = trips_per_month(postCOVID)
```

```python
preCOVID_monthly.plot(title="New York City Taxi trips before and after COVID.", figsize = (7,7), legend = True, ylabel = "Millions of taxi trips")
postCOVID_monthly.plot(xlabel= "Months") # put x-label in the second plot otherwise it will be overwritten
plt.legend(["Trips before COVID", "Trips during COVID"])
```

```python
pre_COVID_avg_passengers = preCOVID['passenger_count'].mean()
post_COVID_avg_passengers = postCOVID['passenger_count'].mean()
```

```python
def average_passengers_per_trip(df):
    # Convert the 'Date' column to a period representing year and month
    df["Year_Month"] = df["Date"].dt.to_period("M")
    
    # Group by 'Year_Month' and calculate total passengers and number of trips
    monthly_data = df.groupby("Year_Month").agg(
        total_passengers=('passenger_count', 'sum'),  # Sum of passengers
        total_trips=('passenger_count', 'size')       # Count of trips
    )
    
    # Calculate average passengers per trip
    monthly_data['average_passengers_per_trip'] = monthly_data['total_passengers'] / monthly_data['total_trips']
    
    # Calculate the overall average
    overall_avg = monthly_data['average_passengers_per_trip'].mean()
    
    return overall_avg, monthly_data['average_passengers_per_trip']

preCOVID_avg, preCOVID_monthly = average_passengers_per_trip(preCOVID)
postCOVID_avg, postCOVID_monthly = average_passengers_per_trip(postCOVID)
```

```python
preCOVID_monthly.plot(title="New York City Taxi average number of passengers per trip before and after COVID.", figsize = (7,7), legend = True, ylabel = "Number of passengers per trip")
postCOVID_monthly.plot(xlabel= "Months")
plt.legend([f"Trips before COVID, average = {pre_COVID_avg_passengers:.2f}", f"Trips during COVID, average = {post_COVID_avg_passengers:.2f}"])
```

```python
trip_counts_preCOVID = preCOVID.groupby(['PULocationID', 'DOLocationID']).size().reset_index(name='trips_preCOVID')
trip_counts_postCOVID = postCOVID.groupby(['PULocationID', 'DOLocationID']).size().reset_index(name='trips_postCOVID')
```

```python
# Create the OD matrix
od_matrix_preCOVID = trip_counts_preCOVID.pivot(index='PULocationID', columns='DOLocationID', values='trips_preCOVID').fillna(0)
od_matrix_postCOVID = trip_counts_postCOVID.pivot(index='PULocationID', columns='DOLocationID', values='trips_postCOVID').fillna(0)

# Convert float to integers for clarity
od_matrix_preCOVID = od_matrix_preCOVID.astype(int)
od_matrix_postCOVID = od_matrix_postCOVID.astype(int)
```
